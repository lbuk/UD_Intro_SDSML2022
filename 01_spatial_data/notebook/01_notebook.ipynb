### Vector
Vector data consists of points, lines and polygons.

#### Vector Geometries
Vector geometries can include Points, LineStrings, Polygons, MultiPoints, MultiLineStrings, MultiPolygons and Geometry Collections. 

We will be utilising Simple Features, a set of standards for spatial information, and WKT (Well Known Text), a text markup language for representing vector geometries, in conjunction with Python.

```python
# Polygon vector geometry
polygon = shapely.wkt.loads('POLYGON ((0 0, 0 -36, 74 -36, 74 0, 0 0))')
```

```python
# Print the data type
type(polygon)
```

```python
# Polygon to Geopandas data frame
polygon_gpd = gpd.GeoSeries(polygon)
```

```python
# Simple map of the polygon
polygon_gpd.plot()
```

```python
# MultiPolygon vector geometry
multipolygon = shapely.wkt.loads('''
                                 MULTIPOLYGON
                                 (((40 35, 20 45, 20 43, 31 20, 40 35)),
                                 ((20 35, 20 30, 10 10, 20 5, 45 20, 20 35)))
                                 ''')
```

```python
# Point vector geometry
point = shapely.geometry.Point((150, 170))
```

```python
# MultiPoint vector geometry
coords = [(2, 1.5), (1.7, 2.2), (0.6, 2), (1, 0), (2.1, 0.5), (1, 1.5)]
multipoint = shapely.geometry.MultiPoint(coords)
```

```python
# Geopandas data frame from the Simple Features MultiPoint
points_gpd = gpd.GeoSeries(multipoint) # Polygon to gpd
```

```python
# Print the data type
type(points_gpd)
```

```python
# Map of the points
points_gpd.plot()
```

```python
# LineString vector geometry
linestring = shapely.geometry.LineString(coords)
```

```python
# MultiLineString vector geometry
linestring_1 = shapely.geometry.LineString([(2, 3.5), (1, 2), (-1.7, 0), (1, -1.1)])
linestring_2 = shapely.geometry.LineString([(-2, 2), (2, -1.5), (2, 0.5)])
multilinestring = shapely.geometry.MultiLineString([linestring_1, linestring_2])
```

```python
# Geometry Collection vector geometry
geometrycollection = shapely.geometry.GeometryCollection([point, multilinestring, multipolygon])
```

```python
# Draw a buffer around the polygon, produce geopandas data frame and visualise
point_buffer = point.buffer(3200)
```

```python
# Polygon buffer to geopandas data frame
point_buffer_gpd = gpd.GeoSeries(point_buffer)
```

```python
# Visualise polygon buffer
point_buffer_gpd.plot()
```

Vector data formats can include shapefiles, GeoJSON, GeoPackage etc. In this example, we will be working with a GeoPackage dataset as well as OpenStreetMap.

```python
# Load the country vector dataset from a Github URL using Geopandas. The dataset is a GeoPackage and was originally sourced from Natural Earth via the Geographic Data Science with Python book: https://geographicdata.science/book/intro.html
country_polygons = gpd.read_file("https://github.com/lbuk/UD_Intro_SDSML2022/blob/master/01_spatial_data/data/world_countries.gpkg?raw=true")
```

```python
# Derive summary info from the dataset
country_polygons.info()
```

```python
# Print the geometry type of the first polygon
type(country_polygons.geometry[0])
```

```python
# Derive the dataset's Coordinate Reference System (CRS)
country_polygons.crs
```

```python
# Print some of the top rows of the dataset
country_polygons.head()
```

```python
# Subset the dataset for Guatemala
guatemala = country_polygons.query('ADMIN == "Guatemala"')
```

```python
# Plot the Guatemala polygon
guatemala.plot()
```

```python
# Load the London land use vector dataset from Github. The dataset is a GeoJSON and was sourced from Corine via data.gov.uk https://www.data.gov.uk/dataset/255d7b61-1a93-4a5c-8d13-2b45952ebf7e/corine-land-cover-2012-for-the-uk-jersey-and-guernsey
landuse = gpd.read_file("https://github.com/lbuk/UD_Intro_SDSML2022/blob/master/01_spatial_data/data/london_landuse.geojson?raw=true")
```

```python
# Derive the dataset info
landuse.info()
```

```python
# Derive the CRS
landuse.crs
```

```python
# Visualise land use in Greater London
landuse.plot("CorineID_L", # Variable to map
             cmap="tab20b", # Qualitative palette
             figsize=(20.7, 20.7), # Map size
             legend=True, # Add a legend
             legend_kwds={
                          "loc": "lower right" # Location of legend
                         }
             )
```

```python
# Summarise landuse by sum of area in hectares
landuse[['AREA_HA', 'CorineID_L']].groupby('CorineID_L').sum()
```

```python
# Summarise landuse by polygon count
landuse[['AREA_HA', 'CorineID_L']].groupby('CorineID_L').count()
```

### OpenStreetMap
[OpenStreetMap](https://www.openstreetmap.org/) is a database of crowdsourced geographic data. We will be using the [OSMnx](https://github.com/gboeing/osmnx) package to source and visualise crowdsourced spatial data from OpenStreetMap.

```python
import osmnx as ox
```

```python
# Geocode Watford using OSMnx
ox.geocoder.geocode_to_gdf("Watford, UK")
```

```python
# Get the OSM network data from Watford using OSMnx
watford_network = ox.graph_from_address('Watford, UK', network_type='drive')
```

```python
type(watford_network)
```

```python
# Visualise the OSM network data from Watford
ox.plot_graph(watford_network)
```

```python
# Calculate the number of graph nodes
len(watford_network.nodes)
```

```python
# Calculate the number of edges from the graph
len(watford_network.edges)
```

```python
# Calculate basic geometrical and topological statistics from the graph
watford_network_stats = ox.basic_stats(watford_network)

# Print basic network stats
# More information about the network stats: https://osmnx.readthedocs.io/en/stable/osmnx.html#module-osmnx.stats
print(watford_network_stats)
```

```python
# Get spatial boundary from OSMnx and produce Geopandas data frame
watford = ox.geocode_to_gdf("Watford, UK")
```

```python
# OSMnx data to Geopandas data frame
watford_gpd = ox.project_gdf(watford)
```

```python
# Map of Watford boundary polygon
watford_gpd.plot()
```

```python
# OSMnx can also be utilised to query OpenStreetMap for building data within an area
watford_buildings = ox.geometries_from_place("Watford, UK", {"building": True})
```

```python
# Visualise OSM buildings in the boundary of Watford
ox.plot_footprints(watford_buildings, color = "white")
```

```python
# Extract the geometry and polygons from the dataset of buildings in Watford
watford_buildings = watford_buildings.loc[:,watford_buildings.columns.str.contains('addr:|geometry')] # Geometry
watford_buildings = watford_buildings.loc[watford_buildings.geometry.type=='Polygon'] # Polygons

# Produce a shapefile with the Watford OSM building data
watford_buildings.to_file('watford_buildings.shp', driver = 'ESRI Shapefile')
```

### Raster
Raster data is made up of continuous surfaces. 

The surfaces of raster datasets are geographically represented as grids with cells and pixels.

```python
# Read in the Greater London population raster dataset using xarray. The data was originally sourced from https://hub.worldpop.org/geodata/summary?id=49903
population_density = xarray.open_rasterio("https://github.com/lbuk/UD_Intro_SDSML2022/blob/master/01_spatial_data/data/london_population.tif?raw=true")
```

```python
population_density.attrs
```

```python
# Plot the raster dataset of population in Greater London. This map shows some errors as the population cannot be less than than zero
population_density.sel(band=1).plot()
```

```python
# Histogram to explore values in raster dataset
population_density.sel(band=1).plot.hist(bins = 8)
```

```python
# Plot where the population in Greater London is more than or equal to zero
population_density.where(population_density >= 0).sel(band=1).plot()
```
