### Vector Data

Vector data is represented by points, lines and polygons.

In this example, we will be working with a vector-based GeoPackage dataset of country polygons.

```python
# Load the dataset from Github using a URL
country_polygons = geopandas.read_file('https://github.com/lbuk/UD_Intro_SDSML2022/blob/master/01_spatial_data/data/countries_clean.gpkg?raw=true')
```

```python
# Derive summary info from the dataset
country_polygons.info()
```

```python
# Print the type of dataset
type(country_polygons.geometry[0])
```

```python
# Print the first few rows of the dataset
country_polygons.head()
```

```python
# Plot the polygons
country_polygons.plot()
```

```python
# To create a GeoJSON file
country_polygons.to_file('countries.geojson', driver="GeoJSON")
```

```python
# Plot the centroids of the country polygons
country_polygons["centroid"] = country_polygons.geometry.centroid
country_polygons.geometry.centroid.plot()
```

```python
# Plot the centroids and shade based on country name
country_polygons.set_geometry("centroid").plot("ADMIN", markersize=5).plot()
```

```python
# Subset the dataset for Brazil
brazil = country_polygons.query('ADMIN == "Brazil"')
```

```python
# Plot the Brazil polygon
brazil.plot()
```

```python
# Produce a GeoJSON file for the Brazil polygon
brazil.to_file('brazil.geojson', driver="GeoJSON")
```

### Networks

Networks are represented as lines. Networks are frequently stored in vector-based datasets.

In this example, we will be using the [OSMnx](https://github.com/gboeing/osmnx) package, which queries OpenStreetMap (OSM) for crowdsourced geographic data.

```python
import osmnx as ox
```

```python
# Geocode Watford
ox.geocoder.geocode_to_gdf("Watford, UK")
```

```python
# Get the OSM network data from Watford. In this example, we have sourced the driving network for Watford
watford_network = ox.graph_from_address('Watford, UK', network_type='drive')
```

```python
type(watford_network)
```

```python
# Visualise the network data from Watford
ox.plot_graph(watford_network)
```

```python
# Calculate the number of graph nodes
len(watford_network.nodes)
```

```python
# Calculate the number of edges from the graph
len(watford_network.edges)
```

```python
# Calculate basic statistics from the graph
watford_network_stats = ox.basic_stats(watford_network)

# Print the circuity average
print(watford_network_stats['circuity_avg']) # Circuity average
```

```python
# Calculate extended stats
ext_watford_network_stats = ox.extended_stats(watford_network, ecc=True, bc=True, cc=True)
```

```python
# Calculate the clustering coefficient average
print(ext_watford_network_stats['clustering_coefficient_avg']) # Betweeness centrality average
```

```python
# OSMnx can also be utilised to query OpenStreetMap for building data
watford_buildings = ox.geometries_from_place("Watford, UK", {"building": True})
```

```python
# Visulise buildings in Watford utilising OSMnx
ox.plot_footprints(watford_buildings, color = "white")
```

```python
# Extract the geometry from the dataset of buildings in Watford
watford_buildings = watford_buildings.loc[:,watford_buildings.columns.str.contains('addr:|geometry')]
watford_buildings = watford_buildings.loc[watford_buildings.geometry.type=='Polygon']

# Produce a shapefile with the Watford OSM building data
watford_buildings.to_file('watford_buildings.shp', driver = 'ESRI Shapefile')
```
