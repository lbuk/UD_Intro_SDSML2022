1. Produce a network visualisation with Python and OSMnx for Ipswich, UK.

```python
# Get the OSM network data from Ipswich
ipswich_network = ox.graph_from_address('Ipswich, UK', network_type='drive')

# Visualise the network data from Ipswich
ox.plot_graph(ipswich_network)
```

2. Utilising OSMnx, download and visualise building data for Ipswich, UK.

```python
# Download OSM building data from Ipswich
ipswich_buildings = ox.geometries_from_place("Ipswich, UK", {"building": True})

# Plot OSM buildings from Ipswich
ox.plot_footprints(ipswich_buildings, color = "white")
```

3. Calculate the number of OSM buildings in Ipswich.

```python
ipswich_buildings = ipswich_buildings.loc[:,ipswich_buildings.columns.str.contains('addr:|geometry')] # OSM geometry
ipswich_buildings = ipswich_buildings.loc[ipswich_buildings.geometry.type=='Polygon'] # OSM Polygons

# Calculate total number of OSM buildings
ipswich_buildings[['geometry']].count()
```

4. Produce a GeoJSON file of the Ipswich building data.

```python
ipswich_buildings.to_file('ipswich_buildings.geojson', driver="GeoJSON") # To GeoJSON
```
