```python
import osmnx as ox
import geopandas
import pandas
import rioxarray
import xarray
import datashader
import contextily as cx
from shapely import geometry
import matplotlib.pyplot as plt
```

1. Produce a network visualisation with Python and OSMnx for Ipswich, UK.

```python
ipswich_network = ox.graph_from_address('Ipswich, UK', network_type='drive')
```

```python
ox.plot_graph(ipswich_network)
```

2. Utilising OSMnx, download and visualise building data for Ipswich, UK.

```python
ipswich_buildings = ox.geometries_from_place("Ipswich, UK", {"building": True})

ox.plot_footprints(ipswich_buildings, color = "white")
```

3. Produce a GeoJSON file of the Ipswich building data.

```python
ipswich_buildings = ipswich_buildings.loc[:,ipswich_buildings.columns.str.contains('addr:|geometry')]
ipswich_buildings = ipswich_buildings.loc[ipswich_buildings.geometry.type=='Polygon']

ipswich_buildings.to_file('ipswich_buildings.geojson', driver="GeoJSON")
```
