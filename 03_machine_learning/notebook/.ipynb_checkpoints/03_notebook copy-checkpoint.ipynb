### Supervised Machine Learning
#### Linear Regression
In supervised machine learning, labelled datasets are utilised to train algorithms. This example will be draw on supervised machine learning techniques, in particular [Ordinary Least Squares (OLS)](https://en.wikipedia.org/wiki/Ordinary_least_squares) spatial and non-spatial linear regression. The data is sourced from the 2019 Indices of Multiple Deprivation (IMD) data from the [London Datastore](https://data.london.gov.uk/dataset/indices-of-deprivation). The dataset contains IMD average scores for Greater London boroughs. To summarise, the higher the average IMD score the more deprived the borough. We can express the simple linear regression equation as:

$$
ln(Y_i) = \alpha + \beta{X_i} + \epsilon_i
$$

in which $ln(Y_i)$ is the logged dependent variable (i.e. IMD average score), $X_i$ is the independent variable (e.g. living environment average score), $\beta$ is the coefficient that give us information about in which way and to what extent each variable is related to the IMD average score, $\alpha$ is the intercept or the expected mean value of $Y_i$ when all $X_i$ equals zero and $\epsilon_i$ is the "error" that captures elements that influence the IMD score but are not included in $X_i$.

```python
# Load the IMD 2019 dataset based on a spreadsheet from the London Datastore: https://data.london.gov.uk/dataset/indices-of-deprivation
imd_2019 = gpd.read_file("~/Downloads/gl_boro_imd_2019.geojson")
```

```python
# Print the dataset info
imd_2019.info
```

```python
# Print the CRS
imd_2019.crs
```

```python
# Specify the dependent variable
# Logging the dependent variable is a way of normalising a skewed variable
imd_2019['log_imd_avg_score'] = np.log(imd_2019['imd_avg_score'])
y = imd_2019['log_imd_avg_score'].values
y_name = 'log_imd_score'
```

```python
# Specify the independent variable
x = np.array([imd_2019.living_env_avg_score]).T
x_name = 'living_env_score'
```

```python
# Produce a scatterplot with IMD average score along the y-axis and living environment average score along the x-axis
fig = plt.figure(figsize=(8.75, 6.5))
ax = sns.regplot(x=x, # x-axis variable
                 y=y, # y-axis variable
                 data=imd_2019, # Data
                 line_kws=dict(color="red")) # Regression line

plt.xlim((15, 45)) # x-axis limits
plt.ylim((0, 3.78)) # y-axis limits

ax.set_xlabel("Living Environment Average Score") # x-axis label
ax.set_ylabel("Log(IMD Average Score)") # y-axis label

ax.set(frame_on=False) # Delete frame

plt.show()
```

A spatial weights matrix connects every borough to its nearest neighbors, which will enable us to add space and get extra diagnostics from the model.

```python
# Create a spatial weights matrix
W = weights.KNN.from_dataframe(imd_2019, k=6)
W.transform = 'r'
```

```python
# Visualise the spatial weights matrix
plot_spatial_weights(W, imd_2019)
```

```python
# Build the non-spatial OLS model
ols = OLS(y=y, # Dependent variable
          x=x, # Independent variable
          w=W, # Spatial weights matrix
          name_y=y_name, # Name of dependent variable
          name_x=[x_name], # Name of independent variable
          name_w="W", # Name of spatial weights matrix
          name_ds='imd_2019', # Name of dataset
          white_test=True, # White’s specification robust test
          spat_diag=True, # Lagrange multiplier tests
          moran=True) # Moran’s I on the residuals

print(ols.summary) # Print a summary of the OLS model
```

This summary of the regression model contains a number of key insights such as the coefficient, the standard error and the p-value. The coefficient is the amount by which change in the independent must be multiplied to give the corresponding average change in the dependent variable, or the amount the dependent variable changes for a unit increase in the independent variable. The standard error represents the average distance that the observed values fall from the regression line, which is shown in the scatterplot. The p-value tests the null hypothesis that the coefficient is equal to zero. The p-value is a statistical measure of the extent to which the independent variable effects the dependent variable.

```python
# Produce a dataset containing the coefficient, standard error and p-value from the OLS model
ols_vals = pandas.DataFrame({"Coeff.": ols.betas.flatten(), # Coefficient
                             "Std. Error": ols.std_err.flatten(), # Standard error
                             "P-Value": [i[1] for i in ols.t_stat], # p-value
                            },index=ols.name_x) # Index of independent variables

print(ols_vals) # Print
```

```python
# Derive the residual values by borough from the OLS model
imd_2019["residual"] = ols.u
```

```python
# Visualise the residuals by borough in a bar chart
fig = plt.figure(figsize=(8.5, 13.75))
ax = plt.gca()

sns.barplot(x="residual", # x-axis variable
            y="local_authority", # y-axis variable
            ax=ax, # Axes
            data=imd_2019.sort_values("residual", ascending = False), # Data, sorted
            color="#d9d9d9") # Palette

ax.set_ylabel('') # Label for y-axis
ax.set_xlabel('Residuals') # Label for x-axis

ax.set(frame_on=False) # Delete frame

plt.show()
```

```python
# Produce a choropleth map of the residuals in the boroughs
fig = plt.figure()
ax = imd_2019.plot(column="residual", # Variable by which to shade
                   figsize=(8.75, 8.75), # Figure size
                   scheme="Quantiles", # Classification scheme
                   cmap="BuPu", # Palette
                   edgecolor='#737373', # Line colour
                   legend=True, # Legend
                   legend_kwds={"loc": "lower right"}) # Legend location

# Font title
font_title = {'family':'helvetica','color':'black','size':14} 
plt.title('Residuals', fontdict=font_title) 

# Annotation of data source
plt.annotate('Data Source: London Datastore', (0,0), (370,-10), fontsize=8, 
             xycoords='axes fraction', textcoords='offset points')

ax.set_axis_off() # Delete axis

ax.add_artist(ScaleBar(1, location='lower left')) # Scale bar

plt.show()
```

```python
# Calculate the spatial lag of the weight and the residual to get the closest IMD average score
knn = weights.KNN.from_dataframe(imd_2019, k=1)
lag_residual = weights.spatial_lag.lag_spatial(knn, ols.u)
```

```python
# Visualise the relationship between the spatial lag and the model residuals utilising a scatterplot with a regression line
fig = plt.figure(figsize=(8.75, 6.5))
ax = sns.regplot(ols.u.flatten(), # x-axis variable
                 lag_residual.flatten(), # y-axis variable
                 line_kws=dict(color="red")) # Regression line

ax.set_xlabel("Model Residuals") # Label of x-axis
ax.set_ylabel("Spatial Lag of Model Residuals") # Label of y-axis

ax.set(frame_on=False) # Delete frame

plt.show()
```

#### Spatially Lagged Exogenous Regressors
In order to introduce space into the regression models, we can lag an independent or explandatory variable. We can express the equation as:

$$
ln(Y_i) = \alpha + \beta X_i + \delta \sum_j w_{ij} X'_i + \epsilon_i
$$

in which $X'_i$ is a subset of $X_i$, although it could encompass all of the independent or explanatory variables, and $w_{ij}$ is the $ij$-th cell of a spatial weights matrix $W$. Because $W$ assigns non-zero values only to spatial neighbors, if $W$ is row-standardized (customary in this context), then $\sum_j w_{ij} X'_i$ captures the average value of $X'_i$ in the surroundings of location $i$. This is what we call the *spatial lag* of $X_i$.

```python
# Produce the spatial weights
w_health_dep_disability_avg_score = weights.KNN.from_dataframe(imd_2019, k=6)

# Assign spatial lag based on the health deprivation and disability average score
lagged = imd_2019.assign(w_health_dep_disability_avg_score=weights.spatial_lag.lag_spatial(w_health_dep_disability_avg_score, imd_2019['health_dep_disability_avg_score'].values))
```

```python
# Add the lagged health deprivation and disability average score variable to the other independent or explanatory variables
explanatory_vars = ['living_env_avg_score', 'health_dep_disability_avg_score']
ext_vars = explanatory_vars + ["w_health_dep_disability_avg_score"]
ext_vars_name = ['living_env_score', 'health_dep_dis_score', 'w_health_dep_dis_score']
```

```python
# Produce the spatial model with the lagged independent or explanatory variable
ols_ii = spreg.OLS(lagged[['log_imd_avg_score']].values, # Dependent variable
                   lagged[ext_vars].values, # Independent variables
                   w=W, # Spatial weights
                   name_y=y_name, # Name of dependent variable
                   name_x=ext_vars_name) # Name of independent variable

print(ols_ii.summary) # Print a summary of the model
```

```python
# Produce a dataset containing the coefficient, standard error and p-value from the model
ols_ii_vals = pandas.DataFrame({"Coeff.": ols_ii.betas.flatten(), # Coefficient
                                "Std. Error": ols_ii.std_err.flatten(), # Standard error
                                "P-Value": [i[1] for i in ols_ii.t_stat], # p-value
                               },index=ols_ii.name_x) # Index of independent variables

print(ols_ii_vals) # Print
```

#### Spatially Lagged Endogenous Regressors
A spatial lag model adds a lag to the dependent variable. Here, we will study how the IMD average score of nearby boroughs impacts the score function utilising a spatial lag model. We can express the equation as:

$$
ln(Y_i) = \alpha + \lambda \sum_j w_{ij} ln(Y_i) + \beta X_i + \epsilon_i
$$

#### Prediction Performance of Models
The spatial models tend to improve prediction accuracy in cases in which space plays a key role in the data generating process. To compare the prediction performance models, we can use the [mean squared error (MSE)](https://en.wikipedia.org/wiki/Mean_squared_error), a standard metric of accuracy in machine learning, to evaluate whether the spatial models are an improvement on the non-spatial model. To summarise, the higher the MSE measure, the higher the error and the worse the performance of the prediction.

```python
explanatory_vars = ['living_env_avg_score', 'health_dep_disability_avg_score']
explanatory_vars_name = ['living_env_score', 'health_dep_dis_score']
```

```python
# Produce the spatial lag model
ols_iii = spreg.GM_Lag(imd_2019[['log_imd_avg_score']].values, # Dependent variable
                       imd_2019[explanatory_vars].values, # Independent variables
                       w=W, # Spatial weights matrix
                       name_y=y_name, # Name of dependent variable
                       name_x=explanatory_vars_name) # Name of independent variables

print(ols_iii.summary) # Print a summary of the model
```

```python
# Calculate the Mean Squared Error (MSE), or prediction performance, of regression models
mses = pandas.Series({'ols': mse(y, ols.predy.flatten()), # ols model (non-spatial)
                      'ols+W': mse(y, ols_ii.predy.flatten()), # ols model with spatially lagged independent or explanatory variable
                      'lag': mse(y, ols_iii.predy.flatten())}) # ols model with spatially lagged dependent variable

mses.sort_values() # Print
```

By including the surrounding health deprivation and disability scores (i.e. ols+W), we have reduced the MSE of the non-spatial model (i.e. ols). The inclusion of the spatial lag of the IMD score (i.e. lag) also improves the accuracy of the non-spatial model (i.e. ols) further.

### Unsupervised Machine Learning
#### Spatial Clustering
In unsupervised machine learning, unlabeled datasets are utilised to identify patterns. In this example we will be drawing on [k-means clustering](https://en.wikipedia.org/wiki/K-means_clustering), an unsupervised machine learning technique. The aforementioned 2019 Index of Multiple Deprivation data on the [London Datastore](https://data.london.gov.uk/dataset/indices-of-deprivation) will be utilised to cluster boroughs in Greater London based on the IMD, living environment and health deprivation and disability average scores.

K-means is one of the most widely utilised clustering techniques. The k-means algorithm groups observations into a pre-specified number of clusters so that each observation is closer to the mean of its own cluster than it is to the mean of any other cluster. First, all observations are randomly assigned one of the 𝑘 labels, although the right number of clusters is typically unknown in practice. Next, the multivariate mean over all covariates is calculated for each of the clusters. Each observation is then reassigned to the cluster with the closest mean.

```python
# Load the IMD 2019 dataset based on a spreadsheet from the London Datastore: https://data.london.gov.uk/dataset/indices-of-deprivation
imd_data = gpd.read_file("~/Downloads/gl_boro_imd_2019.geojson")
```

```python
# Produce a choropleth map of the IMD average score
fig = plt.figure()
ax = imd_data.plot(column="imd_avg_score", # Variable by which to shade by
                   figsize=(8.75, 8.75), # Figure size
                   scheme="Quantiles", # Classification scheme
                   cmap="viridis", # Palette
                   edgecolor='#ffffff', # Line colour
                   legend=True, # Legend
                   legend_kwds={"loc": "lower right"}) # Location of legend

# Font title
font_title = {'family':'helvetica','color':'black','size':14} 
plt.title('IMD Average Score, 2019', fontdict=font_title) 

# Annotation of data source
plt.annotate('Data Source: London Datastore', (0,0), (370,-10), fontsize=8, 
             xycoords='axes fraction', textcoords='offset points')

ax.set_axis_off() # Delete axis

ax.add_artist(ScaleBar(1, location='lower left')) # Scale bar

plt.show()
```

In order to test the extent to which each variable contains spatial structure, we can utilise spatial autocorrelation and Moran’s I, which measures the overall spatial autocorrelation. A positive spatial autocorrelation (or Moran's I) indicates that similar values cluster together and negative spatial autocorrelation for Moran's I occurs when dissimilar values cluster together on a map.

```python
np.random.seed(0)

# Produce spatial weights matrix
w = Queen.from_dataframe(imd_data)

# Set the cluster variables
cluster_variables = ["imd_avg_score",  "living_env_avg_score",  "health_dep_disability_avg_score"]

# Calculate the Moran's I for each variable
mi_results = [
    Moran(imd_data[variable], w) for variable in cluster_variables
]

mi_results = [
    (variable, res.I, res.p_sim)
    for variable, res in zip(cluster_variables, mi_results)
]

# Produce a table of the Moran's I and p-values by variable
morans_i_pval = pandas.DataFrame(
    mi_results, columns=["Variable", "Moran's I", "P-value"]
).set_index("Variable")

print(morans_i_pval) # Print
```

Spatial autocorrelation can only describe relationships between observations and does not indicate how attributes co-vary over space. We can draw on data visualisation to explore this.

```python
# Produce a scatterplot matrix and KDE charts showing the relationships between the variables
fig = plt.figure()
ax = sns.pairplot(imd_data[cluster_variables], # Variables for scatterplot matrix
                  kind="reg", # Scatterplot with regression line
                  diag_kind="kde", # KDE plot
                  aspect=2) # Aspect of chart

ax.set(frame_on=False) # Delete frame

plt.show()
```

When performing spatial clustering, the data should be scaled. Scaling ensures that the data is normalised since distances are sensitive to the units of measurement and cluster solutions can change when you re-scale the data.

```python
# Scale the dataset
imd_data_scaled = robust_scale(imd_data[cluster_variables])
```

```python
# Specify 6 clusters in k-means. The right number of clusters is typically unknown in practice
kmeans = cluster.KMeans(n_clusters=6)
```

```python
# Build the cluster model
kmeans_clusters = kmeans.fit(imd_data_scaled)

# Merge the cluster data with IMD dataset
imd_kmeans_clusters = imd_data.assign(Cluster=kmeans_clusters.labels_)

# Alter formatting so that none of the clusters are equal to zero
imd_kmeans_clusters['Cluster'] = imd_kmeans_clusters['Cluster'] + 1
```

```python
# Produce a choropleth map of clusters
fig = plt.figure()
ax = imd_kmeans_clusters.plot(column='Cluster', # Variable by which to shade by
                              figsize=(8.75, 8.75), # Figure size
                              categorical=True, # Categorical data
                              edgecolor='#ffffff', # Line colour
                              cmap=ListedColormap(['#929084', '#FFC857', '#E5323B', '#A997DF', '#BDD9BF', '#2E4052']), # Palette
                              legend=True, # Legend
                              legend_kwds={"loc": "lower right"}) # Location of legend

# Font title
font_title = {'family':'helvetica','color':'black','size':14}
plt.title('Clusters', fontdict=font_title)

# Annotate data source
plt.annotate('Data Source: London Datastore', (0,0), (370,-10), fontsize=8, 
             xycoords='axes fraction', textcoords='offset points')

ax.set_axis_off() # Delete axis

ax.add_artist(ScaleBar(1, location='lower left')) # Scale bar

plt.show()
```

```python
# Print summary data by cluster and average of the variables
imd_kmeans_clusters.groupby("Cluster")[cluster_variables].mean()
```

```python
# Index dataset on cluster ID
imd_re = imd_kmeans_clusters.set_index("Cluster")

# Subset clustering variables
imd_re = imd_re[cluster_variables]

# Data from wide to long format
imd_re = imd_re.stack()

# Reset index
imd_re = imd_re.reset_index()

# Rename variable and value columns
imd_re = imd_re.rename(columns={"level_1": "Variable", 0: "Value"})
```

```python
# Produce a KDE chart showing the density functions for the clusters and faceted by variable
fig = plt.figure()
ax = sns.FacetGrid(col="Variable", # Variable by which to facet by
                   hue="Cluster", # Variable by which to colour the lines
                   palette=['#929084', '#FFC857', '#E5323B', '#A997DF', '#BDD9BF', '#2E4052'], # Palette
                   data=imd_re, # Dataset
                   sharey=False, # different y-axis for all charts
                   sharex=False, # different x-axis for all charts
                   aspect=1.9, # Chart aspect
                   col_wrap=1) # Number of columns

ax.map(sns.kdeplot, "Value", shade=False).add_legend()

plt.show()
```

#### Google Earth Engine
[Google Earth Engine](https://earthengine.google.com/): "is a cloud-based geospatial analysis platform that enables users to visualise and analyse satellite images of our planet". This example draws on unsupervised machine learning techniques - including clustering - and satellite data to classify land use within Google Earth Engine.

First, the features with numeric properties in which to find clusters are identified. Next, the clusterer is instantiated and the parameters are set. The clusterer is then trained utilising the training data. The clusterer is subsequently applied to a satellite image. Finally, the clusters are labelled in Google Earth Engine.

```python
# conda create -n gee
# conda activate gee
# conda install mamba -c conda-forge
# mamba install geemap -c conda-forge
```

```python
import geemap
import ee
```

```python
# Produce an interactive Google map
Map = geemap.Map()
Map
```

```python
# Generate point data
point = ee.Geometry.Point([-0.1276227, 51.5033725])

# Extract satellite data and filter
image = ee.ImageCollection('LANDSAT/LC08/C01/T1_SR') \
    .filterBounds(point) \
    .filterDate('2019-01-01', '2019-12-31') \
    .sort('CLOUD_COVER') \
    .first() \
    .select('B[1-7]')

# Set visualisation parameters
vis_params = {'min': 0,
              'max': 3000,
              'bands': ['B5', 'B4', 'B3']}

# Map
Map.centerObject(point, 8)
Map.addLayer(image, vis_params, "Landsat-8")
Map
```

```python
# Derive properties from the data
props = geemap.image_props(image)
```

```python
# Print the Landsat ID
props.get('LANDSAT_ID').getInfo()
```

```python
# Print the satellite source
props.get('SATELLITE').getInfo()
```

```python
# Print the cloud cover
props.get('CLOUD_COVER').getInfo()
```

```python
# Build a training dataset and add it to the map
training = image.sample(**{'scale': 30,
                           'numPixels': 5000,
                           'seed': 0,
                           'geometries': True})

# Map
Map.addLayer(training, {}, 'training', False)
Map
```

```python
# Instantiate the clusterer
n_clusters = 6 # Number of clusters
clusterer = ee.Clusterer.wekaKMeans(n_clusters).train(training) # Build the cluster with training data
```

```python
# Cluster the input using the trained clusterer
result = image.cluster(clusterer)
```

```python
# Map
Map.addLayer(result.randomVisualizer(), {}, 'clusters')
Map
```

```python
# Produce and export a tif file of the clusters
geemap.ee_export_image(result, filename='cluster.tif', scale=75)
```
